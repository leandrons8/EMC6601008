# Exerc√≠cio 4

```python
def exercicio4(p1, p2, p3):
    p1 = array(p1)
    p2 = array(p2)
    p3 = array(p3)
    v1 = (p2-p1)/linalg.norm(p2-p1)
    v2 = (p3-p2)/linalg.norm(p3-p2)
    v = array([v1, v2])
    b1 = linalg.norm(p2-p1)/2 + dot(p1, v1)
    b2 = linalg.norm(p3-p2)/2 + dot(p2, v2)
    b = array([b1, b2])
    pc = linalg.solve(v, b)
    
    r = linalg.norm(pc-p1)

    theta = arange(0, 2*pi, 0.01)
    x = pc[0] + r*cos(theta)
    y = pc[1] + r*sin(theta)
    fig = Figure()
    fig.add_trace(
        Scattergl(
            x=x,
            y=y,
            mode='lines',
            name='c√≠rculo'
        )
    )
    fig.add_trace(
        Scatter(
            x=[p1[0], p2[0], p3[0], p1[0]],
            y=[p1[1], p2[1], p3[1], p1[1]],
            text=['P1', 'P2', 'P3'],
            textposition='top center',
            mode='markers+lines+text',
            name='pontos'
        )
    )
    fig.add_trace(
        Scatter(
            x=[pc[0]],
            y=[pc[1]],
            marker_size=10,
            mode='markers',
            name='centro'
        )
    )
    fig.update_yaxes(scaleanchor = "x", scaleratio = 1)
    fig.show()
    return pc


exercicio4(random.rand(2), random.rand(2), random.rand(2))
```

```python
def exercicio5(p1, p2, p3, p4, r):
    p1 = append(array(p1), 0)
    p2 = append(array(p2), 0)
    p3 = append(array(p3), 0)
    p4 = append(array(p4), 0)
    v1 = (p2-p1)/linalg.norm(p2-p1)
    v2 = (p4-p3)/linalg.norm(p4-p3)
    v3 = cross(v1, v2)/linalg.norm(cross(v1, v2))
    v4 = cross(v3, v1)
    v5 = cross(v2, v3)

    t1 = (dot((p3-p1), v5) + (1-dot(v4, v5))*r)/dot((p2-p1), v5)
    t2 = (dot((p1-p3), v4) + (1-dot(v5, v4))*r)/dot((p4-p3), v4)
    pt11 = p1 + t1*(p2-p1)
    pt21 = p3 + t2*(p4-p3)
    pc1 = pt11 + r*v4
    
    u1 = (dot((p3-p1), v5) + (1+dot(v4, v5))*r)/dot((p2-p1), v5)
    u2 = (dot((p1-p3), v4) + (1+dot(v5, v4))*r)/dot((p4-p3), v4)
    pt12 = p1 + u1*(p2-p1)
    pt22 = p3 + u2*(p4-p3)
    pc2 = pt12 - r*v4
    
    v1 = (dot((p3-p1), v5) + (-1-dot(v4, v5))*r)/dot((p2-p1), v5)
    v2 = (dot((p1-p3), v4) + (-1-dot(v5, v4))*r)/dot((p4-p3), v4)
    pt13 = p1 + v1*(p2-p1)
    pt23 = p3 + v2*(p4-p3)
    pc3 = pt13 + r*v4
    
    w1 = (dot((p3-p1), v5) + (-1+dot(v4, v5))*r)/dot((p2-p1), v5)
    w2 = (dot((p1-p3), v4) + (-1+dot(v5, v4))*r)/dot((p4-p3), v4)
    pt14 = p1 + w1*(p2-p1)
    pt24 = p3 + w2*(p4-p3)
    pc4 = pt14 - r*v4

    theta = arange(0, 2*pi, 0.01)
    circulo = [r*cos(theta), r*sin(theta)]
    fig = Figure()
    fig.add_trace(
        Scatter(
            x=pc1[0] + circulo[0],
            y=pc1[1] + circulo[1],
            mode='lines',
            name='c√≠rculo 1'
        )
    )
    fig.add_trace(
        Scatter(
            x=pc2[0] + circulo[0],
            y=pc2[1] + circulo[1],
            mode='lines',
            name='c√≠rculo 2'
        )
    )
    fig.add_trace(
        Scatter(
            x=pc3[0] + circulo[0],
            y=pc3[1] + circulo[1],
            mode='lines',
            name='c√≠rculo 3'
        )
    )
    fig.add_trace(
        Scatter(
            x=pc4[0] + circulo[0],
            y=pc4[1] + circulo[1],
            mode='lines',
            name='c√≠rculo 4'
        )
    )
    fig.add_trace(
        Scatter(
            x=[p1[0], p2[0], p3[0], p4[0]],
            y=[p1[1], p2[1], p3[1], p4[1]],
            text=['P1', 'P2', 'P3', 'P4'],
            textposition='top center',
            marker_size=10,
            mode='markers+text',
            name='pontos'
        )
    )
    fig.add_trace(
        Scatter(
            x=[pc1[0],pc2[0],pc3[0],pc4[0]],
            y=[pc1[1],pc2[1],pc3[1],pc4[1]],
            text=['centro 1', 'centro 2', 'centro 3', 'centro 4'],
            textposition='top center',
            marker_size=10,
            mode='markers+text',
            name='centros'
        )
    )
    fig.add_trace(
        Scatter(
            x=[pt11[0], pt12[0], pt13[0], pt14[0]],
            y=[pt11[1], pt12[1], pt13[1], pt14[1]],
            mode='lines',
            name='reta tangente 1'
        )
    )
    fig.add_trace(
        Scatter(
            x=[pt21[0], pt22[0], pt23[0], pt24[0]],
            y=[pt21[1], pt22[1], pt23[1], pt24[1]],
            mode='lines',
            name='reta tangente 2'
        )
    )
    fig.update_yaxes(scaleanchor = "x", scaleratio = 1)
    fig.show()
    return delete(pc1, 2)


exercicio5(
    random.rand(2),
    random.rand(2),
    random.rand(2),
    random.rand(2),
    random.rand()
)
```

# Exerc√≠cio 7

```python
def exercicio7():
    s, R, b = sp.symbols('s R b')
    r = sp.Matrix(
        [R*sp.cos(s/sp.sqrt(R**2+b**2)),
         R*sp.sin(s/sp.sqrt(R**2+b**2)),
         b*s/sp.sqrt(R**2+b**2)]
    )
    rl = sp.diff(r, s)
    norm_rl = sp.sqrt(rl.dot(rl))
    return sp.simplify(norm_rl)


exercicio7()
```

$$ r(s) = \begin{bmatrix}
R \cos \left( \frac{s}{\sqrt{R^2+b^2}} \right) \\
R \sin \left( \frac{s}{\sqrt{R^2+b^2}} \right) \\
\frac{bs}{\sqrt{R^2+b^2}}
\end{bmatrix} $$

$$ \dot{r(s)}=\begin{bmatrix}
-\frac{R}{\sqrt{R^2+b^2}} \sin \left( \frac{s}{\sqrt{R^2+b^2}} \right) \\
 \frac{R}{\sqrt{R^2+b^2}} \cos \left( \frac{s}{\sqrt{R^2+b^2}} \right) \\
 \frac{b}{\sqrt{R^2+b^2}}
\end{bmatrix} $$

$$ \lVert \dot{r(s)} \rVert = \sqrt{
\left( -\frac{R}{\sqrt{R^2+b^2}} \sin \left( \frac{s}{\sqrt{R^2+b^2}} \right) \right)^2 +
\left(  \frac{R}{\sqrt{R^2+b^2}} \cos \left( \frac{s}{\sqrt{R^2+b^2}} \right) \right)^2 + 
\left(  \frac{b}{\sqrt{R^2+b^2}} \right)^2
} $$

$$ \lVert \dot{r(s)} \rVert = \sqrt{
\frac{R^2}{R^2+b^2} \left( \sin^2 \left( \frac{s}{\sqrt{R^2+b^2}} \right) + \cos^2 \left( \frac{s}{\sqrt{R^2+b^2}} \right) \right) + \frac{b^2}{R^2+b^2}
} $$

$$ \lVert\dot{r(s)}\rVert=\sqrt{\frac{R^2+b^2}{R^2+b^2}} $$

$$ \lVert\dot{r(s)}\rVert=1 $$

```python
def exercicio7(R, b):
    s = arange(0, 10*pi, 0.01)*sqrt(R**2 + b**2)
    r = array(
        [R*cos(s/sqrt(R**2 + b**2)),
         R*sin(s/sqrt(R**2 + b**2)),
         b*s/sqrt(R**2 + b**2)]
    )
    fig = Figure(
        data=Scatter3d(
            x=r[0],
            y=r[1],
            z=r[2],
            mode='lines'
        )
    )
    fig.show()


exercicio7(random.rand(), random.rand())
```

```python
def exercicio8():
    t = linspace(0, 1, 100)
    r = array(
        [3*(t**3)+2*t,
         4*(t**2)+t,
         5*t]
    ).T
    rl = array(
        [9*(t**2)+2,
         8*t+1,
         0*t+5]
    ).T
    rll = array(
        [18*t,
         0*t+8,
         0*t]
    ).T
    rlll = array(
        [0*t+18,
         0*t,
         0*t]
    ).T
    rl_x_rll = cross(rl, rll)
    
    k = linalg.norm(rl_x_rll, axis=1)/(linalg.norm(rl, axis=1)**3)
    tal = sum(rlll * rl_x_rll, axis=1)/(linalg.norm(rl_x_rll, axis=1)**2)
    
    fig = make_subplots(
        rows=1, cols=2, subplot_titles=['curvatura', 'tor√ß√£o'],
        specs=[[{'type': 'scatter3d'}, {'type': 'scatter3d'}]]
    )
    fig.add_trace(
        Scatter3d(
            x=r[:,0],
            y=r[:,1],
            z=r[:,2],
            text=k,
            showlegend=False,
            mode='lines',
            name='curvatura',
            line=scatter3d.Line(
                color=k,
                showscale=True,
                colorbar=scatter3d.line.ColorBar(
                    x=0,
                    xanchor='right'
                )
            )
        ),
        row=1,
        col=1
    )
    fig.add_trace(
        Scatter3d(
            x=r[:,0],
            y=r[:,1],
            z=r[:,2],
            text=tal,
            showlegend=False,
            mode='lines',
            name='tor√ß√£o',
            line=scatter3d.Line(
                color=tal,
                cmid=0,
                colorscale='Portland',
                showscale=True,
                colorbar=scatter3d.line.ColorBar(
                    x=1,
                    xanchor='left'
                )
            )
        ),
        row=1,
        col=2
    )
    fig.show()
    
    fig2 = Figure()
    fig2.add_trace(Scatter(x=t, y=k, name='curvatura'))
    fig2.add_trace(Scatter(x=t, y=tal, name='tor√ß√£o'))
    fig2.show()


exercicio8()
```

```python
def exercicio11(initial_point, final_point, *args):
    """
    initial_point -> [x, y, z, slope]
    final_point -> [x, y, z, slope]
    *args -> [x, y, z, k1, k2]
    """
    a = zeros((2 + 2*len(args), 2 + 2*len(args)))
    b = zeros((2 + 2*len(args), 3))
    a[0,0] = 3
    b[0] = 3*array(initial_point[:3]) + initial_point[3]
    a[-1,-1] = 3
    b[-1] = 3*array(final_point[:3]) - final_point[3]
    p = zeros((2 + len(args), 3))
    p[0] = initial_point[:3]
    p[-1] = final_point[:3]
    v = zeros((4 + 3*len(args), 3))
    v[0] = initial_point[:3]
    v[-1] = final_point[:3]
    for arg in enumerate(args):
        a[2*arg[0]+1, 2*arg[0]+1] = (arg[1][3]+0.5)
        a[2*arg[0]+1, 2*arg[0]+2] = 1
        a[2*arg[0]+2, 2*arg[0]+0] = -(arg[1][4]+0.5)
        a[2*arg[0]+2, 2*arg[0]+1] = 2*(arg[1][4]+0.5)
        a[2*arg[0]+2, 2*arg[0]+2] = -2
        a[2*arg[0]+2, 2*arg[0]+3] = 1
        b[2*arg[0]+1] = array(arg[1][:3])*((arg[1][3]+0.5)+1)
        b[2*arg[0]+2] = array(arg[1][:3])*((arg[1][4]+0.5)-1)
        p[arg[0]+1] = arg[1][:3]
        v[3*arg[0]+3] = arg[1][:3]
    x = linalg.solve(a, b)
    
    counter = 0
    for cp in enumerate(v):
        if cp[0]%3 == 1:
            v[cp[0]] = x[counter]
            counter += 1
        elif cp[0]%3 == 2:
            v[cp[0]] = x[counter]
            counter += 1
        else:
            None
    
    u = linspace(0, 1, 1000)
    curve = []
    for seg in range(len(v)//3):
        r = array(0)
        points = v[3*seg:3*seg+4]
        for i in range(3+1):
            c = 6/(math.factorial(i)*math.factorial(3-i))
            r = r + multiply.outer(c*(u**i)*((1-u)**(3-i)), points[i])
        curve.append(r)
    
    curve = vstack(tuple(curve))
    
    fig = Figure()
    fig.add_trace(
        Scatter3d(
            x=p[:,0],
            y=p[:,1],
            z=p[:,2],
            opacity = 0.5,
            mode='markers',
            name='points'
        )
    )
    fig.add_trace(
        Scatter3d(
            x=v[:,0],
            y=v[:,1],
            z=v[:,2],
            opacity = 0.5,
            name='control points'
        )
    )
    fig.add_trace(
        Scatter3d(
            x=curve[:,0],
            y=curve[:,1],
            z=curve[:,2],
            mode='lines',
            name='curva'
        )
    )
    fig.show()


exercicio11(
    random.rand(4),
    random.rand(4),
    random.rand(5),
    random.rand(5),
    random.rand(5),
    random.rand(5),
    random.rand(5),
)
```

```python
def exercicio13(*args):
    b = array(args)
    n = len(b)
    a = (
        eye(n) +
        eye(n, k=1)*4 +
        eye(n, k=-n+1)*4 +
        eye(n, k=2) +
        eye(n, k=-n+2)
    )/6
    x = linalg.solve(a,b)
    v = vstack((x, x[:3]))
    

    def bspline(v):
        u = linspace(0, 1, 1000)
        us = array([1+u*0, u, u**2, u**3]).T
        mbs = array(
            [[ 1,  4,  1,  0],
             [-3,  0,  3,  0],
             [ 3, -6,  3,  0],
             [-1,  3, -3,  1]]
        )
        v = array(v)
        return matmul(matmul(us, mbs), v)/6
    
    r = zeros(3)
    for n in range(len(v)-3):
        r = vstack((r, bspline(v[n:n+4])))
    r = delete(r, (0), axis=0)

    fig = Figure()
    fig.add_trace(
        Scatter3d(
            x=r[:,0],
            y=r[:,1],
            z=r[:,2],
            mode='lines',
            name='curva'
        )
    )
    fig.add_trace(
        Scatter3d(
            x=b[:,0],
            y=b[:,1],
            z=b[:,2],
            mode='markers',
            name='pontos'
        )
    )
    fig.add_trace(
        Scatter3d(
            x=v[:,0],
            y=v[:,1],
            z=v[:,2],
            name='pontos de controle'
        )
    )
    fig.show()


exercicio13(
    random.rand(3),
    random.rand(3),
    random.rand(3),
    random.rand(3),
    random.rand(3),
)
```

```python
def exercicio14(m):
    i = sp.Symbol('i')
    X = sp.IndexedBase('X')


    def n(m, i, X):
        if m == 1:
            return sp.Piecewise(
                (0, ((X <= X[i]) | (X[i+1] < X))),
                (1, ((X[i] < X) & (X <= X[i+1])))
            )
        else:
            term1 = (X - X[i])/(X[i+m-1] - X[i])*n(m-1, i, X)
            term2 = (X[i+m] - X)/(X[i+m] - X[i+1])*n(m-1, i+1, X)
            return term1+term2
    

    return sp.simplify(n(m, i, X), rational=True)


exercicio14(2)
```

```python
def exercicio15(
    point_zero_plane,
    vector_a_plane,
    vector_b_plane,
    center_sphere,
    radius,
    solid_eq,
    res=50
):
    u, w = meshgrid(linspace(0, 1, res), linspace(0, 1, res))
    plane = array(
        [point_zero_plane[0] + vector_a_plane[0]*u + vector_b_plane[0]*w,
         point_zero_plane[1] + vector_a_plane[1]*u + vector_b_plane[1]*w,
         point_zero_plane[2] + vector_a_plane[2]*u + vector_b_plane[2]*w]
    )
    
    sphere = zeros((3, res, res))
    for i in range(res):
        for j in range(res):
            u, w = i/(res-1), j/(res-1)
            sphere[0, i, j] = radius*sin(pi*u)*cos(2*pi*w) + center_sphere[0]
            sphere[1, i, j] = radius*sin(pi*u)*sin(2*pi*w) + center_sphere[1]
            sphere[2, i, j] = radius*cos(pi*u) + center_sphere[2]
    
    rev = zeros((3, res, res))
    for i in range(res):
        for j in range(res):
            u, w = i/(res-1), j/(res-1)
            rev[0, i, j] = u
            rev[1, i, j] = eval(solid_eq)*cos(4*pi*w)
            rev[2, i, j] = eval(solid_eq)*sin(4*pi*w)
    
    fig = make_subplots(
        rows=1,
        cols=3,
        subplot_titles=['plano', 'esfera', 's√≥lido de revolu√ß√£o'],
        specs=[[
                {'type': 'surface'},
                {'type': 'surface'},
                {'type': 'surface'}]]
    )
    fig.add_trace(
        Surface(x=plane[0], y=plane[1], z=plane[2], showscale=False),
        row=1, col=1
    )
    fig.add_trace(
        Surface(x=sphere[0], y=sphere[1], z=sphere[2], showscale=False),
        row=1, col=2
    )
    fig.add_trace(
        Surface(x=rev[0], y=rev[1], z=rev[2], showscale=False),
        row=1, col=3
    )
    fig.show()


exercicio15(
    point_zero_plane=random.rand(3),
    vector_a_plane=random.rand(3),
    vector_b_plane=random.rand(3),
    center_sphere=random.rand(3),
    radius=random.rand(),
    solid_eq=f'sin(10*pi*u)/10+1',
    res=50
)
```

```python
def exercicio16():
    t, R = sp.symbols('t R')
    u = sp.Function('u')(t)
    w = sp.Function('w')(t)
    r = R*sp.Matrix(
        [sp.sin(sp.pi*u)*sp.cos(2*sp.pi*w),
         sp.sin(sp.pi*u)*sp.sin(2*sp.pi*w),
         sp.cos(sp.pi*u)]
    )
    ru = sp.diff(r, u)
    ruu = sp.diff(ru, u)
    rw = sp.diff(r, w)
    rww = sp.diff(rw, w)
    ruw = sp.diff(ru, w)
    g = sp.Matrix(
        [[ru.dot(ru), ru.dot(rw)],
         [ru.dot(rw), rw.dot(rw)]]
    )
    ru_x_rw = ru.cross(rw)
    n = ru_x_rw/sp.sqrt(ru_x_rw.dot(ru_x_rw))
    d = sp.Matrix(
        [[n.dot(ruu), n.dot(ruw)],
         [n.dot(ruw), n.dot(rww)]]
    )
    vec = sp.Matrix([sp.diff(u, t),sp.diff(w, t)])
    num = vec.T*d*vec
    den = vec.T*g*vec
    return sp.simplify(num[0]/den[0])


exercicio16()
```

# Exerc√≠cio 18

```python
def exercicio18(n, res=50):
    """
    p -> points
    c_p_c_bz -> control points column bezier
    c_p_a_bz -> control points all bezier
    c_p_c_bs -> control points column bspline
    c_p_a_bs -> control points all bspline
    """
    
    
    def bz_c_points(points, random_k):
        """
        b -> 2D array
        i_slope -> 1D array
        f_slope -> 1D array
        k -> 3D array (k[0] = axis, k[:, :, 0] = k1, k[:, :, 1] = k2)
        """
        if random_k:
            kv = random.uniform(0.5, 1, (n-2, 2))
            k = random.choice((True, False), (n-2, 2))
            k = where(k == True, kv, 1/kv)
        else:
            k = ones((n-2, 2))
        b = zeros((3, 2*n-2))
        b[:, 0], b[:, -1] = points[:, 0], points[:, -1]
        for i in range(n-2):
            b[:, 2*i+1] = points[:, i+1]*(k[i, 0]+1)
            b[:, 2*i+2] = points[:, i+1]*(k[i, 0]-1)
        
        a = zeros((2*n-2, 2*n-2))
        a[0, 0], a[-1, -1] = 1, 1
        for i in range(n-2):
            a[2*i+1, 2*i+1] = k[i, 0]
            a[2*i+1, 2*i+2] = 1
            a[2*i+2, 2*i  ] = -k[i, 1]
            a[2*i+2, 2*i+1] = 2*k[i, 1]
            a[2*i+2, 2*i+2] = -2
            a[2*i+2, 2*i+3] = 1
        
        x = linalg.solve(a, b.T).T

        for i in range(n):
            x = insert(x, 3*i, points[:, i], axis=1)
        
        return x
    

    def bs_c_points(points):
        """
        points -> 2D array
        i_slope -> 1D array
        f_slope -> 1D array
        """
        b = zeros((3, n+2))
        b[:, 0], b[:, -1], b[:, 1:-1] = 0, 0, points
        
        a = zeros((n+2,n+2))
        a[0, :3], a[-1, -3:]= array([-3, 0, 3]), array([-3, 0, 3])
        a[1:-1,  :-2] = a[1:-1,  :-2] + eye(n)
        a[1:-1,   2:] = a[1:-1,   2:] + eye(n)
        a[1:-1, 1:-1] = a[1:-1, 1:-1] + 4*eye(n)
        a = a/6
        
        return linalg.solve(a, b.T).T
    

    def get_surf(control_points, surf_type):
        us = array([1 + u*0, u, u**2, u**3])
        if surf_type == 'bezier':
            coef = array(
                [[ 1, 0, 0, 0],
                 [-3, 3, 0, 0],
                 [ 3,-6, 3, 0],
                 [-1, 3,-3, 1]]
            )
            stp = 3
        elif surf_type == 'bspline':
            coef = array(
                [[ 1, 4, 1, 0],
                 [-3, 0, 3, 0],
                 [ 3,-6, 3, 0],
                 [-1, 3,-3, 1]]
            )/6
            stp = 1
        s = zeros((3, (res-1)*(n-1)+1,(res-1)*(n-1)+1))
        for ax in range(3):
            for i in range(n-1):
                for j in range(n-1):
                    s[ax, res*i-i:res*(i+1)-i, res*j-j:res*(j+1)-j] = matmul(
                        matmul(
                            matmul(
                                matmul(
                                    us.T,
                                    coef
                                ),
                                control_points[ax, stp*i:stp*i+4, stp*j:stp*j+4]
                            ),
                            coef.T
                        ),
                        us
                    )
        return s
    

    u = linspace(0, 1, res)
    p=random.rand(3,n,n)
    p[0], p[1] = meshgrid(arange(0,n,1),arange(0,n,1))

    cpcr_bz = zeros((3, 3*n-2, n))
    for i in range(n): cpcr_bz[:, :, i] = bz_c_points(p[:, :, i], True)
    cpar_bz = zeros((3, 3*n-2, 3*n-2))
    for i in range(3*n-2): cpar_bz[:, i, :] = bz_c_points(cpcr_bz[:, i, :], True)
    bzr = get_surf(cpar_bz, 'bezier')

    cpc1_bz = zeros((3, 3*n-2, n))
    for i in range(n): cpc1_bz[:, :, i] = bz_c_points(p[:, :, i], False)
    cpa1_bz = zeros((3, 3*n-2, 3*n-2))
    for i in range(3*n-2): cpa1_bz[:, i, :] = bz_c_points(cpc1_bz[:, i, :], False)
    bz1 = get_surf(cpa1_bz, 'bezier')

    cpc_bs = zeros((3, n+2, n))
    for i in range(n): cpc_bs[:, :, i] = bs_c_points(p[:, :, i])
    cpa_bs = zeros((3, n+2, n+2))
    for i in range(n+2): cpa_bs[:, i, :] = bs_c_points(cpc_bs[:, i, :])
    bs = get_surf(cpa_bs, 'bspline')

    fig = make_subplots(
        rows=1,
        cols=3,
        subplot_titles=[
            'bezier (k1i=rand, k2i=rand)',
            'bezier (k1i=1, k2i=1)',
            'bspline'
        ],
        specs=[[{'type': 'surface'}, {'type': 'surface'}, {'type': 'surface'}]]
    )
    fig.add_trace(
        Scatter3d(
            x=p[0].flatten(),
            y=p[1].flatten(),
            z=p[2].flatten(),
            opacity=0.5,
            mode='markers',
            name='bezier points'
        ),
        row=1,
        col=1
    )
    fig.add_trace(
        Scatter3d(
            x=cpar_bz[0].flatten(),
            y=cpar_bz[1].flatten(),
            z=cpar_bz[2].flatten(),
            opacity=0.5,
            mode='markers',
            name='bezier control points'
        ),
        row=1,
        col=1
    )
    fig.add_trace(
        Surface(
            x=bzr[0],
            y=bzr[1],
            z=bzr[2],
            showscale=False,
            name='bezier surface'
        ),
        row=1,
        col=1
    )
    fig.add_trace(
        Scatter3d(
            x=p[0].flatten(),
            y=p[1].flatten(),
            z=p[2].flatten(),
            opacity=0.5,
            mode='markers',
            name='bezier points'
        ),
        row=1,
        col=2
    )
    fig.add_trace(
        Scatter3d(
            x=cpa1_bz[0].flatten(),
            y=cpa1_bz[1].flatten(),
            z=cpa1_bz[2].flatten(),
            opacity=0.5,
            mode='markers',
            name='bezier control points'
        ),
        row=1,
        col=2
    )
    fig.add_trace(
        Surface(
            x=bz1[0],
            y=bz1[1],
            z=bz1[2],
            showscale=False,
            name='bezier surface'
        ),
        row=1,
        col=2
    )
    fig.add_trace(
        Scatter3d(
            x=p[0].flatten(),
            y=p[1].flatten(),
            z=p[2].flatten(),
            opacity=0.5,
            mode='markers',
            name='bspline points'
        ),
        row=1,
        col=3
    )
    fig.add_trace(
        Scatter3d(
            x=cpa_bs[0].flatten(),
            y=cpa_bs[1].flatten(),
            z=cpa_bs[2].flatten(),
            opacity=0.5,
            mode='markers',
            name='bspline control points'
        ),
        row=1,
        col=3
    )
    fig.add_trace(
        Surface(
            x=bs[0],
            y=bs[1],
            z=bs[2],
            showscale=False,
            name='bspline surface'
        ),
        row=1,
        col=3
    )
    fig.update_layout(showlegend=True)
    fig.show()


exercicio18(10, res=10)
```

# Exerc√≠cio 19

```python
def exercicio19(p, ro, delta):
    u = linspace(0, 1, 1000)
    us = array([1+u*0, u, u**2, u**3])
    mbz = array(
        [[ 1,  0,  0,  0],
         [-3,  3,  0,  0],
         [ 3, -6,  3,  0],
         [-1,  3, -3,  1]]
    )
    v = array(p)
    r = matmul(matmul(us.T, mbz), v)
    
    t = 0
    r_ts = array([p[0]])
    while t < 1:
        rl = matmul(matmul(array([0, 1, 2*t, 3*t**2]), mbz), v)
        rt = matmul(matmul(array([1, t, t**2, t**3]), mbz), v)
        delta_t = 2/linalg.norm(rl)*sqrt(delta*(2*ro-delta))
        t += delta_t
        if t < 1:
            r_ts = vstack((r_ts, rt))
        else:
            r_ts = vstack((r_ts, p[-1]))
    
    fig = Figure()
    fig.add_trace(
        Scatter3d(
            x=v[:,0],
            y=v[:,1],
            z=v[:,2],
            opacity = 0.5,
            name='control points'
        )
    )
    fig.add_trace(
        Scatter3d(
            x=r_ts[:,0],
            y=r_ts[:,1],
            z=r_ts[:,2],
            opacity = 0.5,
            name='ts'
        )
    )
    fig.add_trace(
        Scatter3d(
            x=r[:,0],
            y=r[:,1],
            z=r[:,2],
            mode='lines',
            name='curva'
        )
    )
    fig.show()


exercicio19(random.rand(4,3), 0.1, 0.01)
```
